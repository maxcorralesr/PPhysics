```python
import random
import time

def juego_postapocaliptico_avanzado():
    print("🛑 ¡ALERTA MÁXIMA! 🛑")
    print("El sistema central ha detectado una anomalía.")
    print("Solo tú puedes salvar a la humanidad desactivando el núcleo.")
    print("Primero, selecciona tu nivel de dificultad:")
    print("1. Fácil (90 segundos, pistas precisas)")
    print("2. Medio (60 segundos, pistas normales)")
    print("3. Difícil (45 segundos, pistas vagas)\n")

    # Selección de dificultad
    while True:
        try:
            dificultad = int(input("Elige 1, 2 o 3: "))
            if dificultad not in [1, 2, 3]:
                raise ValueError
            break
        except ValueError:
            print("Opción no válida. Intenta de nuevo.")
    
    # Configurar tiempo y pistas según dificultad
    if dificultad == 1:
        tiempo_limite = 90
        pistas_precisas = True
    elif dificultad == 2:
        tiempo_limite = 60
        pistas_precisas = True
    else:
        tiempo_limite = 45
        pistas_precisas = False

    numero_secreto = random.randint(1, 100)
    intentos = 0
    nivel_estres = 0
    tiempo_inicial = time.time()
    numero_adivinado = False

    print("\n¡La misión comienza ahora! Encuentra el código entre 1 y 100.")
    while not numero_adivinado:
        tiempo_restante = tiempo_limite - (time.time() - tiempo_inicial)
        if tiempo_restante <= 0:
            print("\n💥 BOOM 💥")
            print("El núcleo ha explotado... La humanidad está condenada. 😞")
            print(f"El código secreto era {numero_secreto}.")
            break

        print(f"\n⏰ Tiempo restante: {int(tiempo_restante)} segundos")
        print(f"🧠 Nivel de estrés: {nivel_estres}/100")
        try:
            intento = int(input("🔑 Introduce el código: "))
            
            if intento < 1 or intento > 100:
                print("🚨 Código inválido. Solo números entre 1 y 100.")
                continue
            
            intentos += 1
            nivel_estres += random.randint(5, 15)  # Aumenta el nivel de estrés

            # Comprobar si el código es correcto
            if intento == numero_secreto:
                print("\n🎉 ¡MISIÓN CUMPLIDA! 🎉")
                print(f"Has salvado al mundo en {intentos} intentos. 🌍")
                print("La humanidad te recordará por siempre. ¡Gracias!")
                numero_adivinado = True
            else:
                # Generar pistas dependiendo de la dificultad
                if pistas_precisas:
                    if intento < numero_secreto:
                        print("🔎 El código es MAYOR. ¡Sigue intentando!")
                    else:
                        print("🔎 El código es MENOR. ¡Casi lo logras!")
                else:
                    diferencia = abs(numero_secreto - intento)
                    if diferencia <= 10:
                        print("🔥 Estás muy cerca... ¡No te rindas!")
                    elif diferencia <= 30:
                        print("🌡️ El código está algo lejos, pero no imposible.")
                    else:
                        print("🌀 Estás muy lejos del código. ¿Puedes enfocarte mejor?")
                
                # Simulación de hackeo (distracción con números falsos)
                numeros_falsos = random.sample(range(1, 101), 3)
                print(f"📡 Posibles códigos falsos detectados: {numeros_falsos}")
        
        except ValueError:
            print("🚨 ERROR: Eso no parece un número. ¡Concentra tus esfuerzos!")
        
        # Si el nivel de estrés es crítico
        if nivel_estres >= 100:
            print("\n💣 El estrés te ha superado. No lograste completar la misión. 💀")
            print(f"El código secreto era {numero_secreto}.")
            break

    if not numero_adivinado:
        print("\n👽 *Una voz alienígena resuena*:")
        print("'Sabíamos que los humanos no estaban listos...'")
        print("Fin del juego. 💀")

# Inicia el juego
juego_postapocaliptico_avanzado()
```
