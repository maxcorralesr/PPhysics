```python
import matplotlib.pyplot as plt
import random

# Dimensiones del grid
grid_width = 100
grid_height = 50

# Número de partículas
num_particles = 20

# Generar posiciones aleatorias para las partículas
particles = []
for _ in range(num_particles):
    x = random.uniform(0, grid_width)  # Posición aleatoria en x
    y = random.uniform(0, grid_height)  # Posición aleatoria en y
    particles.append((x, y))

# Dibujar el grid y las partículas
fig, ax = plt.subplots(figsize=(10, 5))

# Dibujar las partículas
for particle in particles:
    circle = plt.Circle(particle, radius=.8, color='blue', alpha=.9)
    ax.add_patch(circle)

# Configurar los límites y el aspecto del grid
ax.set_xlim(0, grid_width)
ax.set_ylim(0, grid_height)
ax.set_aspect('equal', adjustable='box')
plt.grid(visible=True, color='gray', linestyle='--', linewidth=0.5)
plt.title(f"Grid {grid_width}x{grid_height} con {num_particles} partículas")
plt.xlabel("Width")
plt.ylabel("Height")

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dimensiones del grid
grid_width = 100
grid_height = 50

# Número de partículas
num_particles = 20

# Generar posiciones espaciadas linealmente de izquierda a derecha
particles = []
dx = grid_width / (num_particles + 1)  # Espaciado uniforme en x
y_center = grid_height / 2  # Línea central del grid en y

for i in range(1, num_particles + 1):
    x = i * dx  # Distribuir uniformemente a lo largo del ancho
    y = y_center  # Mantener todas las partículas centradas en y
    particles.append((x, y))

# Dibujar el grid y las partículas
fig, ax = plt.subplots(figsize=(10, 5))

# Dibujar las partículas
for particle in particles:
    circle = plt.Circle(particle, radius=0.5, color='blue', alpha=0.7)
    ax.add_patch(circle)

# Configurar los límites y el aspecto del grid
ax.set_xlim(0, grid_width)
ax.set_ylim(0, grid_height)
ax.set_aspect('equal', adjustable='box')
plt.grid(visible=True, color='gray', linestyle='--', linewidth=0.5)
plt.title(f"Grid {grid_width}x{grid_height} con {num_particles} partículas en línea horizontal")
plt.xlabel("Width")
plt.ylabel("Height")

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dimensiones del grid
grid_width = 100
grid_height = 50

# Número de partículas
num_particles = 20

# Generar posiciones distribuidas uniformemente a lo largo del grid
particles = []
rows = int(np.sqrt(num_particles))  # Número de filas aproximado
cols = int(np.ceil(num_particles / rows))  # Número de columnas aproximado

x_spacing = grid_width / (cols + 1)  # Espaciado en x
y_spacing = grid_height / (rows + 1)  # Espaciado en y

for row in range(1, rows + 1):
    for col in range(1, cols + 1):
        if len(particles) < num_particles:
            x = col * x_spacing
            y = row * y_spacing
            particles.append((x, y))

# Dibujar el grid y las partículas
fig, ax = plt.subplots(figsize=(10, 5))

# Dibujar las partículas
for particle in particles:
    circle = plt.Circle(particle, radius=0.5, color='blue', alpha=0.7)
    ax.add_patch(circle)

# Configurar los límites y el aspecto del grid
ax.set_xlim(0, grid_width)
ax.set_ylim(0, grid_height)
ax.set_aspect('equal', adjustable='box')
plt.grid(visible=True, color='gray', linestyle='--', linewidth=0.5)
plt.title(f"Grid {grid_width}x{grid_height} con {num_particles} partículas distribuidas uniformemente")
plt.xlabel("Width")
plt.ylabel("Height")

# Mostrar el gráfico
plt.show()
```
